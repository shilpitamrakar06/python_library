```python
## scalar array (0D) 
l = np.array(8)
print("scalar array=",l)
print("dnim=",l.ndim)
print("shape=",l.shape) 
print("datatype=",l.dtype)
```

```python
## vector array (1D)
l = np.array([1,2,3,4,5]) 
print("vector array=",l)
print("dnim=",l.ndim)
print("shape=",l.shape) 
print("datatype=",l.dtype)
```

```python
## matrix array (2D) 
t = np.array([[1,2,3,4,5,67,7]]) 
print(t.ndim)
print(t.shape)
print(t.dtype)
```
