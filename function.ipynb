```python
import numpy as np
```

## np.zero by default it fills zero 


```python
t1 = np.zeros((3),dtype=np.int32) 
print("1D array")
print(t1)

t2 = np.zeros((3,2),dtype = np.int32) 
print("2D array") 
print(t2) 

t3 = np.zeros((3,3,3),dtype = np.int32) 
print("3D array") 
print(t3)
```

## np.ones() by default it fills one :

```python
t1 = np.ones((3),dtype=np.int32) 
print("1D array")
print(t1)

t2 = np.ones((3,2),dtype = np.int32) 
print("2D array") 
print(t2) 

t3 = np.ones((3,2,3),dtype = np.int32) 
print("3D array") 
print(t3)
```

## np.eye() identity matrix , one on digonal

```python
t = np.eye((3),dtype = np.int32) 
print(t)
```

```python
t = np.eye(2,3)
print(t)
```

## np.empty () random value generate: 

```python
t = np.empty(7) 
print(t)
```

## np.arange() ,(start , end-1 , step) : 

```python
ar = np.arange(1,10) 
print(ar) 
ar = np.arange(1,10,2) 
print(ar)  
ar = np.arange(10,1,-1)  
print(ar)
```
