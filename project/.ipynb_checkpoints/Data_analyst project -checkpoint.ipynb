```python
ax = sns.countplot(x=df['Gender'],data=df,palette=sns.color_palette("Set1"))
plt.title('Comparison between the number of males and females.')
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
data = df.groupby(['Gender'],as_index=False)['Amount'].sum()
data
```

```python
ax=  sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
Amont_Age = df.groupby('Age Group',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
Amont_Age
```

```python
sns.barplot(x='Age Group',y='Amount',data=Amont_Age,palette=sns.color_palette("Set2"))
total = sum(Amont_Age['Amount'])
for i , v in enumerate(Amont_Age['Amount']): 
    per = v / total * 100
    plt.text(i,v,f'{per:.1f}%',ha='center',va='bottom')
plt.show()
```
