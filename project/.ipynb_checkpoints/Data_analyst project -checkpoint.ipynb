```python
ax = sns.countplot(x=df['Gender'],data=df,palette=sns.color_palette("Set1"))
plt.title('Comparison between the number of males and females.')
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
data = df.groupby(['Gender'],as_index=False)['Amount'].sum()
data
```

```python
ax=  sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
Amont_Age = df.groupby('Age Group',as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
Amont_Age
```

```python
sns.barplot(x='Age Group',y='Amount',data=Amont_Age,palette=sns.color_palette("Set2"))
total = sum(Amont_Age['Amount'])
for i , v in enumerate(Amont_Age['Amount']): 
    per = v / total * 100
    plt.text(i,v,f'{per:.1f}%',ha='center',va='bottom')
plt.show()
```

```python
sales_state =df.groupby('State',as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sales_state
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x='State',y='Orders',data=sales_state,palette=sns.color_palette("Set1"))
```

```python
ax = sns.countplot(data=df , x='Marital_Status',palette=sns.color_palette("Set2"))
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
df_marital = df['Marital_Status'].value_counts()
df_marital
```

```python
plt.pie(x=df_marital,labels=['no marital','marital'],autopct='%1.1f%%')
plt.title('Marital Status Distribution')  
plt.show()
```

```python
df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum()
```

```python
sales_Marital_Status = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_Marital_Status
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_Marital_Status)

plt.show()
```

```python
plt.figure(figsize=(16,6))
ax = sns.countplot(data=df , x='Occupation',palette=sns.color_palette("Set2"))
for bars in ax.containers: 
    ax.bar_label(bars)
```

```python
Product_Category_sales = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(20,6))
sns.barplot(data=Product_Category_sales , x='Product_Category',y='Amount',palette=sns.color_palette("Set2"))
plt.show()
```
