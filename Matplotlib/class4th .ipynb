```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import  seaborn as sns 
```

```python
# stacked bar chart == for comparision   
```

```python
countries = ["usa","india","china","japan"] 
gold = [59,43,50,20] 
silver =[120,90,67,82] 
bronze=[224,80,170,120] 
```

```python
w=0.4
h = np.add(gold,silver) 
plt.bar(countries,gold,w,color="yellow",label="gold",ec="black") 
plt.bar(countries,silver,w,color="white",bottom=gold,label="silver",ec="black") 
plt.bar(countries,bronze,w,color="red",bottom=h,label="bronze",ec="black") 
plt.legend()
plt.show()
```
