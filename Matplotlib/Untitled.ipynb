```python
import matplotlib.pyplot as plt  
```

```python
# create first chart 
x = [1,2,3,4,5] 
y=[10,20,30,40,50] 
plt.plot(x,y) 
plt.show()
```

```python
plt.figure(figsize= (4,4))
plt.plot(x,y , color="green" , marker="o",linestyle="dotted",linewidth = 2 ,markersize = 12) 
plt.title("sales with year") 
plt.xlabel("year") 
plt.ylabel("sales")
plt.show()
```

```python
x = [1,2,3,4,5] 
y1=[10,20,30,40,50]
y2 = [20,30,40,50,60] 

plt.figure(figsize=(5,5)) 
plt.plot(x,y1,color="blue", label="sale2024")  
plt.plot(x,y2,color="green",label="sale2025") 
plt.title("sales with year") 
plt.xlabel("year") 
plt.ylabel("sales")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,5))
x = ['A','B','C','D','E'] 
y = [44,67,89,45,20]  
plt.bar(x,y ,color="yellow",ec="black") 
plt.title("sales with year") 
plt.xlabel("year") 
plt.ylabel("sales")
plt.show()
```

```python
# used for distribution analysis 
x = ['A','B','C','D','E'] 
y = [44,67,89,45,20]  
plt.hist(x,bins= 4)
plt.show()
```

```python
## pie chart  is uesd for reletionsal 
categories= ["A",'B','C','D','E'] 
sales= [10,20,30,40,60] 
plt.pie(sales ,labels=categories,autopct="%1.1f%%",startangle=90)
plt.title("pie chart example")
plt.show()
```

```python
## scatterplot is used for reletionship b/w variables 
y1 = [10,20,30,40,50] 
y2 = [67,37,48,57,98] 
plt.scatter(y1,y2) 
plt.show()
```
