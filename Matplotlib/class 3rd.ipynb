```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
state = ["mp","mh","pb","up","ap"] 
area= [200,150,100,75,50] 
s= ["red",'orange','yellow','pink','blue']
plt.bar(state,area,color=s,ec="black",width=0.5)
for i in range (len(state)) : 
    plt.text(i,area[i],area[i],va="bottom",ha="center")  ## va= vertical alignment 
     # xaxis   #yaxis       # print                           ## ha = horizontl alignment 
                                                        
```

```python
year=  [2010,2015,2020,2025,2026] 
scores = [[5,15,6,7,5],[1,7,12,8,6],[2,8,10,11,7],[3,7,11,12,8] ]
x = np.arange(len(year)) 
w=0.2 
plt.bar(x,scores[0],width=w,ec="black",fc="green") 
plt.bar(x+0.20,scores[1],width=w,ec="black",fc="red") 
plt.bar(x+.40,scores[2],width=w,ec="black",fc="yellow") 
plt.bar(x+0.60,scores[3],width=w,ec="black",fc="orange") 
plt.xticks(x,year) 
plt.show
```
