```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
price = np.array([50000,57000,49000,47000,51000]) 
product = np.array([2015,2016,2017,2018,2019]) 
```

```python
plt.figure(figsize=[4,2]) # width,hight 
font1={"family":"arial","color":"red","size":"20"}
font2={"family":"serif","color":"blue","size":"10"}
font3={"family":"comic sans ms","color":"yellow","size":"10"}
plt.plot(product,price) 
plt.title("year vs price",fontdict=font1)
plt.xlabel("year",fontdict=font2) 
plt.ylabel("price",fontdict=font3)
```

```python
plt.figure(figsize=[4,2]) # width,hight 
font1={"family":"arial","color":"red","size":"20"}
font2={"family":"serif","color":"blue","size":"10"}
font3={"family":"comic sans ms","color":"yellow","size":"10"}
plt.plot(product,price,marker="*",ms="10",mec="black",mfc="red",color='yellow') 
plt.title("year vs price",fontdict=font1)
plt.xlabel("year",fontdict=font2) 
plt.ylabel("price",fontdict=font3)
```

```python

df = pd.read_csv("cricket.csv")
df
```

```python
plt.figure(figsize=[10,5]) 
plt.plot(df["Year"],df["Rohit"],label='rohit',linestyle="dotted",linewidth = "7") 
plt.plot(df["Year"],df["Virat"],label='virat',linestyle="dashed") 
plt.plot(df["Year"],df["Dhoni"],label="dhoni") 
plt.legend()
plt.savefig("cricket")  ## save as image : 
plt.show()
```
