```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('Customer Churn.csv') 
df
```

```python
df.info()
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['totalcharges'] = df['totalcharges'].replace(" ","0")
df['totalcharges'] = df['totalcharges'].astype("float") 
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df['customerid'].duplicated().sum()
```

```python
df['seniorcitizen'] = df['seniorcitizen'].replace({0:"no",1:"yes"})
```

```python
df.sample(40)
```

```python
ax=sns.countplot(x='churn',data=df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(4,5))
gb = df.groupby('churn').agg({"churn":'count'}) 
e= [0.1,0]
plt.pie(gb['churn'],labels=gb.index ,autopct="%1.2f%%",explode=e,shadow=True)
plt.title("percentage of churn customers")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x="gender", data=df, hue="churn") 
plt.title("churn by gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x="seniorcitizen", data=df) 
ax.bar_label(ax.containers[0])
plt.title("churn by seniorcitizen")
plt.show()
```

```python
##   comparative a greater number of people in sseniorcitizen category have churned 
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x="tenure",data=df,hue="churn",bins=72) 

plt.show()
```

```python
## people who have used  our services for a long time have  stayed and people have used  # 1 or 2 months have churned out 
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(x="contract", data=df,hue="churn") 
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("churn by contract")
plt.show()
```

```python
##  people who have month to month contract  are likely  to chrun then from those who have 1 or 2 year of contract 
```

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(x="paymentmethod", data=df,hue="churn") 
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("churn by paymentmethod")
plt.xticks(rotation=45)
plt.show()
```
