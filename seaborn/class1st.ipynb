```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset("flights") 
data
```

```python
dt = sns.load_dataset("tips") 
dt
```

```python
## scatterplot == use for linear growth ,axis level function 
```

```python
sns.scatterplot(dt,x="total_bill",y="day")
```

```python
plt.figure(figsize=[10,5])
sns.scatterplot(dt,x="total_bill",y="tip",hue="sex",style="smoker",size="size") 
plt.show
```

```python
sns.lineplot(dt,x="total_bill",y="tip")
plt.show()
```

```python
## relational plot (figure level function) 
#1scatter plot 
## line plot 
```

```python
sns.relplot(dt,x="total_bill",y="tip",kind="line") 
sns.relplot(dt,x="total_bill",y="tip",kind="scatter") 
plt.show
```

```python
## A Facetgrid in seaborn (also know as a trellis or small- multiple plot ) 
# allowa split your data 
# into multiple subplots based on categorical variables 
g= sns.FacetGrid(dt,col="sex",hue="day") 
g
```

```python
g.map(plt.scatter,"total_bill","tip").add_legend()
plt.show()
```
