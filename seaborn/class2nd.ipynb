```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
## matrix plot 
#heatmap :: use  for numerical column 
#cluster 
```

```python
## heatmap 
dt = sns.load_dataset("tips")
dt.head() 
dt.columns
t = dt[["total_bill","tip","size"]].head(10)
t
sns.heatmap(t.corr(),annot=True) ## corr :: for relation 
```

```python
##cluster map  
pt =sns.load_dataset("iris") 
pt
sns.clustermap(pt.iloc[:,[0,1,2,3]]) 
plt.show()
```
