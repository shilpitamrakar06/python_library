```python
import pandas as pd 
import numpy as np
```

```python
a = [1,2,3,4,5] 
print(a)
print(type(a))
k = pd.Series(a) 
print(k) 
print(type(k))
```

```python
a = (1,2,3,4,5,6,7,8) 
print(a) 
print(type(a))
k = pd.Series(a) 
print(k)
print(type(k))
```

```python
a = {'x':1,'y':2,'z':3} 
print(a)
print(type(a))
k = pd.Series(a)
print(k)
print(type(k))
```

```python
## python is a object oriented programming language : 
```

```python
a = [1,2,3,"hello",True,None] 
print(a) 
print(type(a))
k = pd.Series(a) 
print(k) 
print(type(k))
```

```python
a = [1,2,3,"hello",True]  
k = pd.Series(a,name="cybrom") 
print(k.name)
print(k.is_unique)
```

```python
a = [1,2,3,4,5,5,5,'5',67] 
k = pd.Series(a)
print(k)
print(k.is_unique)
```

```python
a = [1,2,3,'4',5,6,7] 
k = pd.Series(a)
print(k)
print(k.is_unique)
```

```python
a = [True,None] 
k = pd.Series(a) 
print(k.is_unique)
```

```python
a = [1,2,3,'4',5,6,7] 
k = pd.Series(a)
print(k)
print(k.index)
print(k.values)
```

```python
a = [0,2,4,6,8]
k = pd.Series(a) 
print(k)
print(k.index)
```

```python
a = {'name':['a','b'],'age':[19,20]} 
k = pd.Series(a)
print(k)
print(k.index)
print(k['name'][1])
```

```python
a = [1,2,3,4,5] 
S = pd.Series(a,index=['a','b','c','d','e']) 
print(S)
print(S.index)
print(S.values)
```
