```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('myrgroup.csv') 
print(df.isnull().sum())
print(df)
print(df.dropna()) 
```

```python
A = [[1,2,3,np.nan],[np.nan,np.nan,np.nan , np.nan],[10,20,30,40]] 
df1 = pd.DataFrame(A) 
df1.dropna(how= 'all')  ## drop whole row   
```

```python
A = [[1,2,3],[4,5,6,7,7],[2,3,4,5]] 
df = pd.DataFrame(A) 
df
df.fillna(2)  ## replace none values 
```

```python
# fillna : used to fill a value in every none value in dataframe

A = [[1,2,3,4,5],[6,7,8,9,0],[None,None,None,None,None]]
a = pd.DataFrame(A, columns=['s1','s2','s3','s4','s5'])
# a.fillna(0)         default filling
print(a['s3'].fillna(5))     # filling at a specific position
# a['s3'] = a['s3'].fillna(5)    for permanent change
print(a)
print(a['s4'].fillna(a['s4'].mean()))    # replacing with mean value
a['s5'].ffill() 

# ffill fill  forword 
# bfill backword fill 
 
```
