```python
import pandas as pd 
import numpy as np
```

```python
name = ['phy','chem','math','eng'] 
score = [78,65,89,65] 
s = pd.Series(score,index=name) 
print(s)
```

```python
score = [1,2,3,4,5,6,7,8,9,0,2,3,4,5,6,7,8,9,12,3,4,5,6,7,8,9,0,67,37,28,19,26,37,45,56,35] 
s = pd.Series(score) 
print(s)
print(s.head())# by default give 5 record top 5 
print(s.head(15))
print(s.tail())# by default give 5 record bottom 5 
print(s.tail(12))
print(s.sample())# by default give 1 record . 
print(s.sample(4))
```

```python
score = [1,2,3,4,5,6,7,8,9,0,2,3,4,5,6,7,8,9,12,3,4,5,6,7,8,9,0,67,37,28,19,26,37,45,56,35] 
s = pd.Series(score) 
print(s.value_counts())
```

```python
s = ['a','p','p','l','e'] 
a = pd.Series(s) 
print(a.value_counts())
print(a.sort_values())
print(a.sort_values(ascending=False))
```
