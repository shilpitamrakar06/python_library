```python
import pandas as pd 
import numpy as np 
```

```python
## for even columns: 

df1 = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]}) 
df2 = pd.DataFrame({"C":[10,20,30,40],"D":[50,60,70,80]}) 
print(df1) 
print(df2)
```

```python
df1.join (df2) ## join two table  
```

```python
# for uneven columns : it fills nan 

df1 = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]}) 
df2 = pd.DataFrame({"C":[10,20],"D":[50,60]}) 
df1.join (df2)
```

```python
# for uneven columns: assigning index for uneven columns : it wont disply  the ddata of less 
df1 = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]},index=['a','b','c','d']) 
df2 = pd.DataFrame({"C":[10,20],"D":[50,60]}) 
df1.join(df2) ## priority with left dataframe 
```

```python
df1 = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]},index=['a','b','c','d']) 
df2 = pd.DataFrame({"C":[10,20],"D":[50,60]},index= ['c','d'])  
df1.join(df2) 
```

```python
df1 = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]}) 
df2 = pd.DataFrame({"C":[10,20,30,40],"D":[50,60,70,80]},index= ['a','b','c','d'])   
df2.join(df1)
```

```python
df3 = pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]},index=['a','b','c','d']) 
df4 = pd.DataFrame({"C":[10,20,30,40],"D":[50,60,70,80]},index= ['a','b','x','d'])   
df1.join(df2)  
print("inner join",df3.join(df4,how="inner")) 
print("left join",df3.join(df4,how="left"))
print("right join",df3.join(df4,how="right")) 
```
