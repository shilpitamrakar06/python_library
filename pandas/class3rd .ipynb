```python
import pandas as pd 
import numpy as np
```

```python
## describe function = five number function (five values ) 
a = [10,20,30,40,50,60,70,80] 
s = pd.Series(a) 
print(s)
print(s.describe())
```

```python
## fancy index[[]] randam indexing value 

a = [10,20,30,40,50,60,70,80] 
s = pd.Series(a) 
s[[0,3,4]] = 100
print(s)
```

```python
## null variation :
# isnull() :# retun true\false 

a = [10,20,np.nan,40,50,60,70,80,np.nan,None]  
s = pd.Series(a) 
#print(s)
print(s.isnull().sum()) 
# hasnan :# return only true 
print(s.hasnans)
# notnull : opsite isnull 
print(s.notnull())
```

```python
a = [1,2,3,4,5,3,4,5,27,8,9,] 
s = pd.Series(a) 
print(s)
print(s.count())
print(s.unique())
print(s.nunique())
```

```python
a = [2,4,7,8,3,5,6,7]
s = pd.Series(a)
print(s)
print(sorted(s)) # sort to asecnding 

print(sorted(s)[::-1]) ## sort to descending 

print(len(s))
print(type(s))
print(dir(s))
print(max(s))
```
