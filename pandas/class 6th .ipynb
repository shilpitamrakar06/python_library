```python
import pandas as pd 
import numpy as np 
```

```python
data = {"A":[1,2,3,4],"B":[4,4,7,8]} 
df = pd.DataFrame(data) 
print(df) 
print(df.sum())
print(df["A"].sum()) 
print(df.max()) 
print(df["B"].max()) 
print(df.min()) 
print(df["A"].min()) 
print(df.mean()) 
print(df["A"].mean()) 
print(df.mode()) 
print(df['B'].mode()) 
print(df.var()) 
print(df["A"].var()) 
print(df.std()) 
print(df["B"].std())
```
